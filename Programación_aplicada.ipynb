## Limpieza de datos

```python
import pandas as pd

# Inicializar una lista para las columnas a eliminar
cols_to_drop = []

# Iterar sobre las columnas del DataFrame
for col in datos.columns:
    # Convertir los valores de la columna a texto
    col_values = datos[col].astype(str)

    # Verificar si "TOTAL" está en alguna celda de la columna
    if col_values.str.contains('TOTAL', case=False, na=False).any():
        cols_to_drop.append(col)

# Eliminar las columnas identificadas
datos_cleaned = datos.drop(columns=cols_to_drop)

# Lista de nombres para eliminar
nombres_a_eliminar = [
    "Amazonas", "Antioquia", "Arauca", "Archipiélago de San Andrés, Providencia y Santa Catalina",
    "Atlántico", "Bogotá, D.C.", "Bolívar", "Boyacá", "Caldas", "Caquetá", "Casanare", "Cauca",
    "Cesar", "Chocó", "Córdoba", "Cundinamarca", "Guainía", "Guaviare", "Huila", "La Guajira",
    "Magdalena", "Meta", "Nariño", "Norte de Santander", "Putumayo", "Quindío", "Santander",
    "Sucre", "Tolima", "Valle del Cauca", "Vaupés", "Vichada"
]

# Inicializar una lista para las filas a eliminar
indices_a_eliminar = []

# Iterar sobre las filas del DataFrame
for i in range(len(datos_cleaned)):
    # Verificar si alguna celda en la columna 2 contiene un nombre a eliminar
    if datos_cleaned.iloc[i, 1] in nombres_a_eliminar:
        indices_a_eliminar.append(i)

# Eliminar las filas identificadas
datos_cleaned = datos_cleaned.drop(indices_a_eliminar).reset_index(drop=True)

# Eliminar las últimas 4 columnas
datos_cleaned = datos_cleaned.iloc[:, :-4]

# Reiniciar el índice
datos_cleaned.reset_index(drop=True, inplace=True)

# Establecer la primera fila como los nombres de las columnas
datos_cleaned.columns = datos_cleaned.iloc[0]  # Establece la primera fila como encabezado
datos_cleaned = datos_cleaned[1:]  # Elimina la primera fila del DataFrame
datos_cleaned.reset_index(drop=True, inplace=True)  # Reiniciar el índice

# Crear DataFrames para las columnas específicas
hombre = datos_cleaned.iloc[:, [0, 1, 2, 3]]  # Columnas 0, 1, 2
hombre.columns = hombre.iloc[0]
mujer = datos_cleaned.iloc[:, [0, 1, 4, 5]]  # Columnas 0, 3, 4
mujer.columns = mujer.iloc[0]
hombre = hombre[1:]
mujer = mujer[1:]

# Cambiar el nombre de la columna "N°" a "Divipola"
hombre.rename(columns={"N°": "Divipola"}, inplace=True)
mujer.rename(columns={"N°": "Divipola"}, inplace=True)

#Cambiar nombre Municipio
hombre.rename(columns={"DEPARTAMENTO Y MUNICIPIO DEL HECHO DANE / MAYOR DE EDAD (> 18 AÑOS) O MENOR DE EDAD (< 18 AÑOS)": "Municipio"}, inplace=True)
mujer.rename(columns={"DEPARTAMENTO Y MUNICIPIO DEL HECHO DANE / MAYOR DE EDAD (> 18 AÑOS) O MENOR DE EDAD (< 18 AÑOS)": "Municipio"}, inplace=True)
```

## Anulación de dinamización de columnas

```python
# Función para combinar las columnas de edad en un formato deseado
def reshape_age_columns(df):
    # Crear una lista para almacenar las filas nuevas
    new_rows = []

    # Iterar sobre cada fila del DataFrame original
    for _, row in df.iterrows():
        # Iterar sobre las columnas de edad
        for col in ['Menores de Edad (<18 años)', 'Mayores de Edad (>18 años)']:
            if pd.notna(row[col]):  # Verificar si el valor no es NaN
                new_row = {
                    'Divipola': row['Divipola'],  # Mantener el valor de 'Divipola'
                    'Municipio': row['Municipio'],  # Mantener el nombre
                    'Edad': col,  # El nombre de la columna original
                    'Casos': row[col]  # El valor de la columna actual para 'Casos'
                }
                new_rows.append(new_row)

    # Crear un nuevo DataFrame con las filas transformadas
    df_reshaped = pd.DataFrame(new_rows)

    return df_reshaped

# Aplicar la función al DataFrame
hombre = reshape_age_columns(hombre)
mujer = reshape_age_columns(mujer)
```

## Unir dfs

```python
# Agregar una nueva columna 'Género' con el valor constante 'hombre' o 'mujer'
hombre['Género'] = 'hombre'
mujer['Género'] = 'mujer'
```

```python
# Unir los DataFrames `hombre` y `mujer` verticalmente
data_unida = pd.concat([hombre, mujer], ignore_index=True)
data_unida["Año"] = 2022
```

# Para 2019 a 2022

```python
import pandas as pd

años = [2019, 2020, 2021, 2022]
data_f = []

for año in años:
    # Leer el archivo Excel
    datos = pd.read_excel(f'/content/5-Violencia interpersonal. Colombia, {año}.xlsx', sheet_name="DEPARTAMENTO Y MUNICIPIO")

    # Eliminar las últimas 7 filas
    datos = datos.iloc[:-7]

    # Inicializar una lista para las columnas a eliminar
    cols_to_drop = []
    for col in datos.columns:
        col_values = datos[col].astype(str)
        if col_values.str.contains('TOTAL', case=False, na=False).any():
            cols_to_drop.append(col)

    # Eliminar las columnas identificadas
    datos_cleaned = datos.drop(columns=cols_to_drop)

    # Lista de nombres para eliminar
    nombres_a_eliminar = [
        "Amazonas", "Antioquia", "Arauca", "Archipiélago de San Andrés, Providencia y Santa Catalina",
        "Atlántico", "Bogotá, D.C.", "Bolívar", "Boyacá", "Caldas", "Caquetá", "Casanare", "Cauca",
        "Cesar", "Chocó", "Córdoba", "Cundinamarca", "Guainía", "Guaviare", "Huila", "La Guajira",
        "Magdalena", "Meta", "Nariño", "Norte de Santander", "Putumayo", "Quindío", "Santander",
        "Sucre", "Tolima", "Valle del Cauca", "Vaupés", "Vichada"
    ]

    # Filtrar filas que contienen nombres a eliminar en cualquier celda
    datos_cleaned = datos_cleaned[~datos_cleaned.apply(lambda row: row.astype(str).isin(nombres_a_eliminar).any(), axis=1)]

    # Eliminar las últimas 4 columnas
    datos_cleaned = datos_cleaned.iloc[:, :-4]

    # Establecer la primera fila como los nombres de las columnas
    datos_cleaned.columns = datos_cleaned.iloc[0]
    datos_cleaned = datos_cleaned[1:]
    datos_cleaned.reset_index(drop=True, inplace=True)

    # Crear DataFrames para las columnas específicas
    hombre = datos_cleaned.iloc[:, [0, 1, 2, 3]]  # Columnas 0, 1, 2, 3
    hombre.columns = hombre.iloc[0]
    hombre = hombre[1:]
    hombre.reset_index(drop=True, inplace=True)

    mujer = datos_cleaned.iloc[:, [0, 1, 4, 5]]  # Columnas 0, 1, 4, 5
    mujer.columns = mujer.iloc[0]
    mujer = mujer[1:]
    mujer.reset_index(drop=True, inplace=True)

    # Cambiar el nombre de las columnas
    hombre.rename(columns={"N°": "Divipola"}, inplace=True)
    mujer.rename(columns={"N°": "Divipola"}, inplace=True)
    hombre.rename(columns={"DEPARTAMENTO Y MUNICIPIO DEL HECHO DANE / MAYOR DE EDAD (> 18 AÑOS) O MENOR DE EDAD (< 18 AÑOS)": "Municipio"}, inplace=True)
    mujer.rename(columns={"DEPARTAMENTO Y MUNICIPIO DEL HECHO DANE / MAYOR DE EDAD (> 18 AÑOS) O MENOR DE EDAD (< 18 AÑOS)": "Municipio"}, inplace=True)

    # Función para combinar las columnas de edad en un formato deseado
    def reshape_age_columns(df):
        new_rows = []
        for _, row in df.iterrows():
            for col in ['Menores de Edad (<18 años)', 'Mayores de Edad (>18 años)']:
                if pd.notna(row[col]):
                    new_row = {
                        'Divipola': row['Divipola'],
                        'Municipio': row['Municipio'],
                        'Edad': col,
                        'Casos': row[col]
                    }
                    new_rows.append(new_row)
        return pd.DataFrame(new_rows)

    # Aplicar la función a los DataFrames
    hombre = reshape_age_columns(hombre)
    mujer = reshape_age_columns(mujer)

    # Agregar una nueva columna 'Género'
    hombre['Género'] = 'hombre'
    mujer['Género'] = 'mujer'

    # Unir los DataFrames `hombre` y `mujer`
    data_unida = pd.concat([hombre, mujer], ignore_index=True)
    data_unida["Año"] = año

    # Añadir el DataFrame del año a la lista
    data_f.append(data_unida)

# Concatenar todos los DataFrames en la lista
data_final_2019_2022 = pd.concat(data_f, ignore_index=True)


# Opcional: Guardar el DataFrame final en un archivo Excel
data_final_2019_2022.to_excel('/content/data_final_2019_2022.xlsx', index=False)
```

# Años 2017 y 2018

```python
for año in [2017, 2018]:
  # Leer el archivo Excel
  datos = pd.read_excel(f'/content/5-Violencia interpersonal. Colombia, {año}.xlsx', sheet_name="DEPARTAMENTO Y MUNICIPIO")
  # Eliminar las últimas 7 filas
  datos = datos.iloc[1:-7, :-2]

  # Lista de nombres para eliminar
  nombres_a_eliminar = [
      "Amazonas", "Antioquia", "Arauca", "Archipiélago de San Andrés, Providencia y Santa Catalina",
      "Atlántico", "Bogotá, D.C.", "Bolívar", "Boyacá", "Caldas", "Caquetá", "Casanare", "Cauca",
      "Cesar", "Chocó", "Córdoba", "Cundinamarca", "Guainía", "Guaviare", "Huila", "La Guajira",
      "Magdalena", "Meta", "Nariño", "Norte de Santander", "Putumayo", "Quindío", "Santander",
      "Sucre", "Tolima", "Valle del Cauca", "Vaupés", "Vichada"
  ]

  # Filtrar filas que contienen nombres a eliminar en cualquier celda
  datos_cleaned = datos[~datos.apply(lambda row: row.astype(str).isin(nombres_a_eliminar).any(), axis=1)]
  datos_cleaned.reset_index(drop=True, inplace=True)

  # Crear DataFrames para las columnas específicas
  hombre = datos_cleaned.iloc[:, [0, 1, 2]]  # Columnas 0, 1, 2, 3
  hombre.reset_index(drop=True, inplace=True)

  mujer = datos_cleaned.iloc[:, [0, 1, 4]]  # Columnas 0, 1, 4, 5
  mujer.reset_index(drop=True, inplace=True)

  #Cambiar nombre a columna 2
  hombre.rename(columns={f"VIOLENCIA INTERPERSONAL EN COLOMBIA AÑO {año} / DEPARTAMENTO, MUNICIPIO DEL HECHO DANE": "Municipio"}, inplace=True)
  mujer.rename(columns={f"VIOLENCIA INTERPERSONAL EN COLOMBIA AÑO {año} / DEPARTAMENTO, MUNICIPIO DEL HECHO DANE": "Municipio"}, inplace=True)

  #Cambiar nombre 3ra columna
  hombre.rename(columns={"VIOLENCIA INTERPERSONAL": "Casos"}, inplace=True)
  mujer.rename(columns={"VIOLENCIA INTERPERSONAL.2": "Casos"}, inplace=True)

  # Cambiar el nombre de la primera columna a 'Divipola'
  hombre.rename(columns={hombre.columns[0]: "Divipola"}, inplace=True)
  mujer.rename(columns={mujer.columns[0]: "Divipola"}, inplace=True)

  # Unir los DataFrames `hombre` y `mujer`
  data_unida = pd.concat([hombre, mujer], ignore_index=True)
  data_unida["Año"] = año

  data_2017_2018 = data_unida
```
